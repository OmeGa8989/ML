```python
import pandas as pd
from sklearn.datasets import load_iris 
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head(5)
```

```python
df['target']=iris.target
df.head()
```

```python
iris.target_names
# 0 for setosa
# 1 for versicolor
# 3 for virginica 
# setting target values for classification of the flowers 
```

```python
df[df.target==3].head()
df[df.target==4].head()

# here we can see that only 0 , 1 ,2  are the classification targets(verified).
```

```python
df ['flowerName'] = df.target.apply(lambda x : iris.target_names[x]);
df.head()
```
